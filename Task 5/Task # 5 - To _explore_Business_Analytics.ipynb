## 4.3   profit Analysis

```python
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
sns.barplot(x='Ship Mode', y="Profit", data=df)
plt.subplot(1,2,2)
sns.barplot(x='Ship Mode', y="Sales", data=df)
```

- here, we observed that, ame day sales have more Profit

```python
tab=df.groupby(['State']).Sales.sum().reset_index()
tab.sort_values(by=['Sales'], inplace=True)
tab.head(15)
```

- Above Are Less Sales generated states

```python
tab=df.groupby(['State']).Profit.sum().reset_index()
tab.sort_values(by=['Profit'], inplace=True)
tab.head(15)
```

- Texas,Ohio,Pennsylvania,Illinois,North Carolina, Colorado,Tennessee,Arizona,Florida,Oregon are losse incured states

```python
plt.figure(figsize=(25,20))
sns.barplot(y='State', x="Profit", data=df)
```

```python
df.groupby(['Region']).Profit.sum().reset_index()
```

```python
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
sns.barplot(x='Region', y="Profit", data=df)
plt.subplot(1,2,2)
sns.barplot(x='Region', y="Sales", data=df)
```

```python
df.groupby(['Segment']).Profit.sum().reset_index()
```

```python
sns.barplot(x='Segment', y="Profit", data=df)
```

```python
df.groupby(['City']).Profit.sum().reset_index()
```

```python
tab=df.groupby(['City']).Profit.sum().reset_index()
tab.sort_values(by=['Profit'], inplace=True)
tab.head(15)
```

- Above are 15 most loss incurred city

```python
plt.figure(figsize=(25,20))
sns.barplot(x='City', y="Profit", data=df)
```
